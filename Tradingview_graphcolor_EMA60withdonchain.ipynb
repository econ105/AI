```python
//@version=5
indicator("Close - EMA60 with Donchian (300)", overlay = false)

// Inputs
lenEma   = input.int(60, "EMA Length")
lenDonch = input.int(300, "Donchian Length")

// EMA60
ema60 = ta.ema(close, lenEma)

// Indicator line: close minus EMA60
diff = close - ema60

// Donchian channel on the indicator line over past 300 bars
upperDiff = ta.highest(diff, lenDonch)
lowerDiff = ta.lowest(diff, lenDonch)

// Zero line (EMA60 baseline of the indicator)
hline(0, "Zero", color = color.gray, linestyle = hline.style_dotted)

// Plot main line
pDiff  = plot(diff, color = color.new(color.blue, 0), linewidth = 2, title = "Close - EMA60")

// Plot Donchian channel lines
pHigh  = plot(upperDiff, color = color.new(color.green, 0), linewidth = 1, title = "Donchian High")
pLow   = plot(lowerDiff, color = color.new(color.red, 0), linewidth = 1, title = "Donchian Low")

// Optional fill between channel lines
fill(pHigh, pLow, color = color.new(color.green, 90))

// --- Labels at last bar ---
if barstate.islast
    label.new(bar_index, 0,
         text = "EMA60: " + str.tostring(ema60, format.mintick),
         xloc = xloc.bar_index,
         yloc = yloc.price,
         style = label.style_label_left,
         color = color.new(color.gray, 0),
         textcolor = color.white)

    upperPrice = ema60 + upperDiff
    label.new(bar_index, upperDiff,
         text = "EMA60+Upper: " + str.tostring(upperPrice, format.mintick),
         xloc = xloc.bar_index,
         yloc = yloc.price,
         style = label.style_label_left,
         color = color.new(color.green, 0),
         textcolor = color.white)

    lowerPrice = ema60 + lowerDiff
    label.new(bar_index, lowerDiff,
         text = "EMA60+Lower: " + str.tostring(lowerPrice, format.mintick),
         xloc = xloc.bar_index,
         yloc = yloc.price,
         style = label.style_label_left,
         color = color.new(color.red, 0),
         textcolor = color.white)
```
