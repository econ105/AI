```python
# Install required libraries
!pip install PyMuPDF python-docx

from google.colab import files
import fitz  # PyMuPDF
from docx import Document

# Function to upload PDF
def upload_pdf():
    print("Please upload your PDF file.")
    uploaded = files.upload()
    pdf_filename = list(uploaded.keys())[0]
    return pdf_filename

# Upload PDF
pdf_filename = upload_pdf()

# Function to convert PDF to Word
def pdf_to_word(pdf_filename):
    doc = Document()
    pdf_document = fitz.open(pdf_filename)

    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        text = page.get_text("text")

        # Clean and structure the text
        lines = text.split('\n')
        paragraph = []

        for line in lines:
            if line.strip():
                if len(paragraph) > 0 and line[0].isupper():
                    doc.add_paragraph(' '.join(paragraph))
                    paragraph = [line.strip()]
                else:
                    paragraph.append(line.strip())
            else:
                if paragraph:
                    doc.add_paragraph(' '.join(paragraph))
                    paragraph = []

        if paragraph:
            doc.add_paragraph(' '.join(paragraph))

    word_filename = pdf_filename.replace('.pdf', '.docx')
    doc.save(word_filename)
    return word_filename

# Convert the uploaded PDF to Word
word_filename = pdf_to_word(pdf_filename)
print(f"Converted {pdf_filename} to {word_filename}")

# Download the Word Document
files.download(word_filename)
```
