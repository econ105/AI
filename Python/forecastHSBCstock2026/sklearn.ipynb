```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Historical HSBC HK data (Jan 2026 sample)
data = {
    "Date": [
        "9/1/2026", "10/1/2026", "11/1/2026", "12/1/2026", "15/1/2026",
        "16/1/2026", "17/1/2026", "18/1/2026", "19/1/2026", "22/1/2026",
        "23/1/2026", "24/1/2026", "25/1/2026", "26/1/2026"
    ],
    "Open": [
        134.2, 133.8, 132.5, 133.0, 132.0,
        131.5, 130.8, 131.2, 130.0, 129.5,
        130.2, 131.0, 132.1, 131.8
    ]
}

# Step 2: Create DataFrame
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df = df.sort_values(by="Date").reset_index(drop=True)

# Step 3: Define prediction dates (next 5 trading days)
prediction_data = pd.DataFrame({
    "Date": pd.to_datetime(
        ["27/1/2026", "28/1/2026", "29/1/2026", "30/1/2026", "31/1/2026"],
        dayfirst=True
    )
})

# Step 4: Train linear regression model
model = LinearRegression()
model.fit(df.index.values.reshape(-1, 1), df["Open"])

# Step 5: Predict next 5 days
prediction_data["Open"] = model.predict(prediction_data.index.values.reshape(-1, 1))

# Step 6: Display forecast
print(prediction_data)
```
