```python
# Install & extract (same as before)
!pip install yt-dlp pandas openpyxl
import pandas as pd
from yt_dlp import YoutubeDL
import os
from google.colab import files  # For Colab download

video_url = "https://www.youtube.com/watch?v=P1ikZ8Dic20"

opts = {'getcomments': True, 'writecomments': False}
with YoutubeDL(opts) as yt:
    info = yt.extract_info(video_url, download=False)
    comments = info.get('comments', [])
    print(f"Extracted {len(comments)} comments.")

# Create Excel in notebook filesystem
rows = [{'author': c.get('author'), 'text': c.get('text'),
         'likes': c.get('like_count', 0), 'time': c.get('time_text')} for c in comments]
df = pd.DataFrame(rows)
filename = 'ai_comments.xlsx'
df.to_excel(filename, index=False, engine='openpyxl')

# DOWNLOAD IMMEDIATELY to your computer
files.download(filename)
print("Downloaded ai_comments.xlsx to your Downloads folder!")
```
