```python
from google.colab import files
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster

# Upload the file
uploaded = files.upload()

# Load the CSV data into a DataFrame
df = pd.read_csv(next(iter(uploaded)))

# Check for and handle missing values
print("Missing values in the DataFrame:")
print(df[['LATITUDE', 'LONGITUDE']].isnull().sum())

# Drop rows with NaN values in 'LATITUDE' or 'LONGITUDE'
df.dropna(subset=['LATITUDE', 'LONGITUDE'], inplace=True)

# Check data structure after dropping NaNs
print("Data after dropping NaNs:")
print(df.head())

# Create a GeoDataFrame
df['LATITUDE'] = df['LATITUDE'].astype(float)
df['LONGITUDE'] = df['LONGITUDE'].astype(float)

# Create a GeoDataFrame from the DataFrame
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.LONGITUDE, df.LATITUDE))

# Plotting with Folium
m = folium.Map(location=[gdf['LATITUDE'].mean(), gdf['LONGITUDE'].mean()], zoom_start=12)

# Adding points to the map with a marker cluster
marker_cluster = MarkerCluster().add_to(m)

for idx, row in gdf.iterrows():
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        popup=f"{row['ADDRESS']}, {row['CITY']}, {row['STATE OR PROVINCE']}: ${row['PRICE']}",
    ).add_to(marker_cluster)

# Save the map to an HTML file
map_file = "property_map.html"
m.save(map_file)

print(f"Map has been saved to {map_file}.")
```
