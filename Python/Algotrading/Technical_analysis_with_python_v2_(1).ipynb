## Crossover/crossunder

```python
df["EMA20"]
```

```python
df["EMA20"].shift()
```

```python
import pandas as pd

# --- Step 1: Inspect column structure ---
print("Columns:", df.columns)

# --- Step 2: Normalize EMA columns ---
# Ensure EMA20 and EMA60 are Series, not DataFrames
if isinstance(df["EMA20"], pd.DataFrame):
    ema20 = df["EMA20"].iloc[:, 0]   # take first sub-column
else:
    ema20 = df["EMA20"]

if isinstance(df["EMA60"], pd.DataFrame):
    ema60 = df["EMA60"].iloc[:, 0]
else:
    ema60 = df["EMA60"]

# --- Step 3: Define crossover and crossunder conditions ---
df["EMA_crossover"] = (ema20 > ema60) & (ema20.shift() <= ema60.shift())
df["EMA_crossunder"] = (ema20 < ema60) & (ema20.shift() >= ema60.shift())

# --- Step 4: Optional: mark signals visually ---
df["Signal"] = 0
df.loc[df["EMA_crossover"], "Signal"] = 1   # Buy signal
df.loc[df["EMA_crossunder"], "Signal"] = -1 # Sell signal

# --- Step 5: Quick check ---
print(df[["EMA20", "EMA60", "EMA_crossover", "EMA_crossunder", "Signal"]].tail())
```

## Gap

```python
import pandas as pd

# --- Step 1: Inspect column structure ---
print("Columns:", df.columns)

# --- Step 2: Normalize columns into Series ---
def get_series(df, colname):
    """Return a Series even if df[colname] is a DataFrame."""
    col = df[colname]
    return col.iloc[:,0] if isinstance(col, pd.DataFrame) else col

ema20 = get_series(df, "EMA20")
ema60 = get_series(df, "EMA60")
open_ = get_series(df, "Open")
close = get_series(df, "Close")

# --- Step 3: Compute crossover & crossunder ---
df["EMA_crossover"] = (ema20 > ema60) & (ema20.shift() <= ema60.shift())
df["EMA_crossunder"] = (ema20 < ema60) & (ema20.shift() >= ema60.shift())

# --- Step 4: Compute gap (open vs prior close) ---
df["Gap"] = open_ - close.shift()

# --- Step 5: Optional unified signal column ---
df["Signal"] = 0
df.loc[df["EMA_crossover"], "Signal"] = 1   # Buy
df.loc[df["EMA_crossunder"], "Signal"] = -1 # Sell

# --- Step 6: Quick check ---
print(df[["Open","Close","EMA20","EMA60","Gap","EMA_crossover","EMA_crossunder","Signal"]].tail())
```
