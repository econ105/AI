```python
import numpy as np
import matplotlib.pyplot as plt

# Quantity of units produced
Q = np.linspace(0, 100, 200)

# Define cost function: TC = Fixed Cost + Variable Cost
fixed_cost = 200
variable_cost_per_unit = 20
quadratic_term = 0.1  # increasing marginal cost

total_cost = fixed_cost + variable_cost_per_unit * Q + quadratic_term * Q**2

# Define revenue function: TR = Price * Quantity
price_per_unit = 50
total_revenue = price_per_unit * Q

# Plotting
plt.figure(figsize=(10,6))
plt.plot(Q, total_cost, label="Total Cost", color="red")
plt.plot(Q, total_revenue, label="Total Revenue", color="blue")

# Mark break-even point(s)
idx = np.argwhere(np.diff(np.sign(total_revenue - total_cost))).flatten()
for i in idx:
    plt.scatter(Q[i], total_cost[i], color="black", zorder=5)
    plt.text(Q[i], total_cost[i], f"Break-even at Q={Q[i]:.0f}", fontsize=9)

plt.title("Total Cost and Total Revenue Curves")
plt.xlabel("Quantity Produced")
plt.ylabel("Cost / Revenue")
plt.legend()
plt.grid(True)
plt.show()
```
